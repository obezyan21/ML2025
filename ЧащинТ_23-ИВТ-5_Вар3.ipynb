Отмена рейсов

Задание 1

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=df, x='CANCELLED', hue='CANCELLED', discrete=True, stat='count')
plt.title('Распределение отменённых и не отменённых рейсов')
plt.xlabel('Не отменён (0) / Отменён (1)')
plt.ylabel('Количество рейсов')
plt.show()
```

Задание 2

```python
cancel_by_airport = df[df['CANCELLED'] == 1].groupby('ORIGIN_AIRPORT').size().reset_index(name='Cancelled_Flights')
cancel_by_airport = cancel_by_airport.sort_values('Cancelled_Flights', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=cancel_by_airport, x='Cancelled_Flights', y='ORIGIN_AIRPORT', hue='ORIGIN_AIRPORT', palette='Reds_r')
plt.title('Топ-10 аэропортов по отменённым рейсам')
plt.xlabel('Количество отмен')
plt.ylabel('Аэропорт')
plt.show()

worst_airport = cancel_by_airport.iloc[0]['ORIGIN_AIRPORT']
print(f"Аэропорт с наибольшим числом отмен: {worst_airport}")
```

Задание 3

```python
from scipy.stats import pearsonr, spearmanr

pearson_day, p_day = pearsonr(df['DAY_OF_WEEK'], df['CANCELLED'])
spearman_day, _ = spearmanr(df['DAY_OF_WEEK'], df['CANCELLED'])
pearson_month, p_month = pearsonr(df['MONTH'], df['CANCELLED'])
spearman_month, _ = spearmanr(df['MONTH'], df['CANCELLED'])

print(f"Пирсон (день недели): {pearson_day:.3f}, p-value: {p_day:.3f}")
print(f"Спирмен (день недели): {spearman_day:.3f}")
print(f"Пирсон (месяц): {pearson_month:.3f}, p-value: {p_month:.3f}")
print(f"Спирмен (месяц): {spearman_month:.3f}")

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.kdeplot(data=df, x='MONTH', hue='CANCELLED', common_norm=False, legend = False)
plt.title('Распределение отмен по месяцам')

plt.subplot(1, 2, 2)
sns.kdeplot(data=df, x='DAY_OF_WEEK', hue='CANCELLED', common_norm=False, legend = False)
plt.title('Распределение отмен по дням недели')
plt.show()
```

Задание 4

```python
airlines = ['HA', 'MQ', 'VX']

for airline in airlines:
    airline_data = df[df['AIRLINE'] == airline]
    total_flights = len(airline_data)
    cancelled = airline_data['CANCELLED'].sum()
    diverted = airline_data['DIVERTED'].sum()

    print(f"\nАвиакомпания: {airline}")
    print(f"Всего рейсов: {total_flights}")
    print(f"Отменено: {cancelled}")
    print(f"Перенаправлено: {diverted}")

    sns.histplot(data=airline_data, x='CANCELLED', hue='CANCELLED', discrete=True, legend=False)
    plt.title(f'Отмены рейсов ({airline})')
    plt.show()
```

Задание 5

```python
df['SPEED'] = df['DISTANCE'] / (df['AIR_TIME'] / 60)

for airline in airlines:
    speed_data = df[df['AIRLINE'] == airline]['SPEED'].dropna()
    print(f"\nАвиакомпания: {airline}, Средняя скорость: {speed_data.mean():.2f} км/ч")
    sns.histplot(speed_data, kde=True)
    plt.title(f'Распределение скорости ({airline})')
    plt.show()
```

Задание 6

```python
sample_days = df['DAY'].drop_duplicates().sample(10, random_state=42)
sample_airports = df['ORIGIN_AIRPORT'].drop_duplicates().sample(2, random_state=42)

heatmap_data = df[
    (df['DAY'].isin(sample_days)) &
    (df['ORIGIN_AIRPORT'].isin(sample_airports))
].pivot_table(
    index='FLIGHT_NUMBER',
    columns='DAY',
    values='CANCELLED',
    aggfunc='sum',
    fill_value=0
)

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='YlOrRd', annot=True, fmt='g')
plt.title('Тепловая карта отмен по дням и рейсам')
plt.xlabel('День месяца')
plt.ylabel('Номер рейса')
plt.show()
```

Задание 7

```python
delays = df.groupby('AIRLINE').agg({
    'DEPARTURE_DELAY': 'mean',
    'ARRIVAL_DELAY': 'mean'
}).reset_index()

print(delays.sort_values('DEPARTURE_DELAY', ascending=False))
```

Задание 8

```python
top = delays.sort_values('DEPARTURE_DELAY', ascending=False).head(5)['AIRLINE']

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data=df[df['AIRLINE'].isin(top)], x='DEPARTURE_DELAY', hue='AIRLINE', kde=True)
plt.title('Задержки отправления')

plt.subplot(1, 2, 2)
sns.histplot(data=df[df['AIRLINE'].isin(top)], x='ARRIVAL_DELAY', hue='AIRLINE', kde=True)
plt.title('Задержки прибытия')
plt.show()
```

Задание 9

```python
worst_dep = delays.loc[delays['DEPARTURE_DELAY'].idxmax()]['AIRLINE']
worst_arr = delays.loc[delays['ARRIVAL_DELAY'].idxmax()]['AIRLINE']

print(f"Наибольшие задержки отправления: {worst_dep}")
print(f"Наибольшие задержки прибытия: {worst_arr}")
```
